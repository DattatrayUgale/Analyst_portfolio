```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dff=pd.read_csv(r'D:\DMU\Data_Analyst_Portfolio\Netflix Data Analysis\netflix_titles.csv')
dff.shape
```

```python
dff.columns
```

```python
a = dff["rating"].value_counts().head()
index = a.index
rating = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(5, 5))
plt.pie(rating, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of Content Ratings on Netflix", fontsize=20)
plt.show()
```
